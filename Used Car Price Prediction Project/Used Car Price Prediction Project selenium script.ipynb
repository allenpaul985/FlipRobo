```python
brand=[]
model=[]
variant=[]
manufacturingyear=[]
driverkilometers=[]
fuel=[]
owners=[]
location=[]
price=[]



for i in url:
    driver.get(i)
    sleep(6)
    

#Brand   
    try:
        br=driver.find_element_by_xpath("//span[@data-aut-id='value_make']")
        brand.append(br.text)
    except NoSuchElementException:
        brand.append('-')

#Model        
    try:
        mo=driver.find_element_by_xpath("//span[@data-aut-id='value_model']")
        model.append(mo.text)
    except NoSuchElementException:
        model.append('-')
        
#Variant
    try:
        var=driver.find_element_by_xpath("//span[@data-aut-id='value_variant']")
        variant.append(var.text)
    except NoSuchElementException:
        variant.append('-')

#Manufacturingyear
    try:
        yr=driver.find_element_by_xpath("//span[@data-aut-id='value_year']")
        manufacturingyear.append(yr.text)
    except NoSuchElementException:
        manufacturingyear.append('-')

#Kilometers Driven
    try:
        dr=driver.find_element_by_xpath("//span[@data-aut-id='value_mileage']")
        driverkilometers.append(dr.text)
    except NoSuchElementException:
        driverkilometers.append('-')
        
#Fuel
    try:
        fl=driver.find_element_by_xpath("//span[@data-aut-id='value_petrol']")
        fuel.append(fl.text)
    except NoSuchElementException:
        fuel.append('-')

#no of owners
    try:
        own=driver.find_element_by_xpath("//span[@data-aut-id='value_first_owner']")
        owners.append(own.text)
    except NoSuchElementException:
        owners.append('-')
        
#location
    try:
        loc=driver.find_element_by_xpath("(//div[@class='fontweight500 value'])[1]")
        location.append(loc.text)
    except NoSuchElementException:
        location.append('-')
        
#Price
    try:
        pr=driver.find_element_by_xpath("//span[@class='_2xKfz']")
        price.append(pr.text)
    except NoSuchElementException:
        price.append('-')




brand = pd.DataFrame(brand)            
brand.columns=['brand']

model = pd.DataFrame(model)            
model.columns=['model']

variant = pd.DataFrame(variant)            
variant.columns=['variant']

manufacturingyear = pd.DataFrame(manufacturingyear)            
manufacturingyear.columns=['manufacturingyear']

driverkilometers = pd.DataFrame(driverkilometers)            
driverkilometers.columns=['driverkilometers']

fuel = pd.DataFrame(fuel)            
fuel.columns=['fuel']

owners = pd.DataFrame(owners)            
owners.columns=['owners']

location = pd.DataFrame(location)            
location.columns=['location']

price = pd.DataFrame(price)            
price.columns=['price']



car=pd.DataFrame({})
car['Brand']=brand['brand']
car['Model']=model['model']
car['Variant']=variant['variant']
car['Manufacturig year']=manufacturingyear['manufacturingyear']
car['Kilometers driven']=driverkilometers['driverkilometers']
car['Fuel']=fuel['fuel']
car['No of Owners']=owners['owners']
car['Location']=location['location']
car['Price']=price['price']
car
```
