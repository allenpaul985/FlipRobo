```python
#intel core 17

filter_button=driver.find_elements_by_xpath("(//span[@class='a-size-base a-color-base'])")
for i in filter_button:
    if i.text=='Intel Core i7':
        i.click()
        break
```

```python
laptitle=[]
for j in driver.find_elements_by_xpath("(//span[@class='a-size-medium a-color-base a-text-normal'])"):
    if len(laptitle)<10:
        laptitle.append(j.text)

laptitle = pd.DataFrame(laptitle)
laptitle.columns=['laptitle']
laptitle
#top 10 laptops for i7 in amazon.com
```

```python
lapprice=[]
for j in driver.find_elements_by_xpath("(//span[@class='a-price-whole'])"):
    if len(lapprice)<10:
        lapprice.append(j.text)

lapprice = pd.DataFrame(lapprice)
lapprice.columns=['lapprice']
lapprice
#top 10 laptop's price for i7 in amazon.com
```

```python
ratingurli7=[] #scraping all urls of first 10 i7 laptops
for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
    if len(ratingurli7)<10:
        ratingurli7.append(i.get_attribute('href'))
ratingurli7
```

```python
lapratingi7=[]
for i in ratingurli7:
    url=i
    driver.get(url)
    sleep(4)
    try:
        ratingi7=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        lapratingi7.append(ratingi7.text)
    except NoSuchElementException:
        lapratingi7.append('None')
lapratingi7 = pd.DataFrame(lapratingi7)            
lapratingi7.columns=['laprating']
lapratingi7
```

```python
laptopi7=pd.DataFrame({})
laptopi7['Title']=laptitle['laptitle']
laptopi7['Price']=lapprice['lapprice']
laptopi7['Rating']=lapratingi7['laprating']
laptopi7
#first 10 i7 laptop's Title, price and ratings
```
