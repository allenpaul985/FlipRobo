from www.amazon.in. The product verticals to be searched will be taken as input from user.

2. In the above question, now scrape the following details of each product listed in first 3 pages 
of your search results and save it in a dataframe and csv. In case if any product vertical has 
less than 3 pages in search results then scrape all the products available under that product 
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and “Product URL”. In case, if any of the details are missing for any of the product then

```python
producturl=[] #scraping all urls of first 10 i7 laptops
for i in range(0,3):
    for j in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a"):
                producturl.append(j.get_attribute('href'))
    driver.find_element_by_xpath("//li[@class='a-last']").click()
    sleep(10)
producturl
```

```python
brandname=[]
productname=[]
rating=[]
noofratings=[]
price=[]
Return=[]
delivery=[]
availability=[]
return_exchange=[]
otherdetails=[]


for i in producturl:
    url=i
    driver.get(url)
    sleep(10)
    

#productname   
    try:
        prod=driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        productname.append(prod.text)
    except NoSuchElementException:
        productname.append('-')

#brandname        
    try:
        brad=driver.find_element_by_xpath("//td[@class='a-span9'][1]")
        brandname.append(brad.text)
    except NoSuchElementException:
        brad=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/a")
        brandname.append(brad.text)
        
#rating
    try:
        ratingi7=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(ratingi7.text)
    except NoSuchElementException:
        rating.append('-')

#no.of ratings
    try:
        noofrat=driver.find_element_by_id('acrCustomerReviewText')
        noofratings.append(noofrat.text)
    except NoSuchElementException:
        noofratings.append('-')

#price
    try:
        pric=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        price.append(pric.text)
    except NoSuchElementException:
        price.append('-')

#Return
    try:
        retrn=driver.find_element_by_xpath("(//a[@class='a-size-small a-link-normal a-text-normal'])[1]")
        return_exchange.append(retrn.text)
    except NoSuchElementException:
        return_exchange.append('-')
#delivery
    try:
        deliv=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b")
        delivery.append(deliv.text)
    except NoSuchElementException:
        delivery.append('-')

#availability
    try:
        avail=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
        availability.append(avail.text)
    except NoSuchElementException:
        availability.append('-')

        
#other details

    for i in driver.find_elements_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']"):
        otherdetails.append(i.text)



brandname = pd.DataFrame(brandname)            
brandname.columns=['brandname']

productname = pd.DataFrame(productname)            
productname.columns=['productname']

rating = pd.DataFrame(rating)            
rating.columns=['rating']

noofratings = pd.DataFrame(noofratings)            
noofratings.columns=['noofratings']

price = pd.DataFrame(price)            
price.columns=['price']

return_exchange = pd.DataFrame(return_exchange)            
return_exchange.columns=['return_exchange']

delivery = pd.DataFrame(delivery)            
delivery.columns=['delivery']

availability = pd.DataFrame(availability)            
availability.columns=['availability']

otherdetails = pd.DataFrame(otherdetails)            
otherdetails.columns=['otherdetails']
otherdetails['otherdetails']=otherdetails['otherdetails'].str.replace('\n',' ')


producturl = pd.DataFrame(producturl)            
producturl.columns=['producturl']


samsungphones=pd.DataFrame({})
samsungphones['Brand Name']=brandname['brandname']
samsungphones['Name of the Product']=productname['productname']
samsungphones['Rating']=rating['rating']
samsungphones['No of Ratings']=noofratings['noofratings']
samsungphones['Price']=price['price']
samsungphones["Return/Exchange"]=return_exchange['return_exchange']
samsungphones['Expected Delivery']=delivery['delivery']
samsungphones['Availability']=availability['availability']
samsungphones['Other Details']=otherdetails['otherdetails']
samsungphones['Product URL']=producturl['producturl']
samsungphones
```
